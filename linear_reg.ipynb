```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df=pd.read_csv(r"/content/data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.corr()
```

```python
x = df.Duration
y = df.Calories
```

```python
m,c,rvalue,pvalue,stderr = stats.linregress(x,y)
```

```python
def predictiveline(x):
  return m * x + c
```

```python
predy = list(map(predictiveline,x))
```

```python
plt.scatter(x,y)
plt.plot(x,predy)
plt.show()
```
